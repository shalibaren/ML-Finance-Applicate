```python
from opencc import OpenCC# (Optional )convert from Simplified Chinese to Traditional Chinese
#1-9.txt檔為我們自己標註的訓練集
"""
標註規則與範例如下:

規則1:
文章內容....+(/o空格)機構(/nt空格) +...文章內容....+(/o空格)人名(/nr空格)+...文章內容....+(/o空格)地點(/ns空格)+...文章內容....+(/o空格)換行 

例如:
2003年/o 台北地院/nt 將/o 章民強/nr 等人/o 判處無罪/o 


規則2:
機構(/nt空格) +...文章內容....+(/o空格)人名(/nr空格)+...+(/o空格)換行

例如:
衛福部立台東醫院/nt 已故前精神科主任/o 陳仁文/nr 涉詐領健保、收取藥商回扣等案/o 

"""
news_data=[]
for i in range(1,9):
    fp = open('./data_raw/'+str(i)+'.txt','r')# 有時候編碼有錯，要加,encoding="utf-8"
    to_convert = fp.read()
    to_convert = to_convert.replace("\n", '')
    cc = OpenCC('t2s')  # (Optional )convert from Simplified Chinese to Traditional Chinese
    content = cc.convert(to_convert)

    data_list = list(content.strip().split(' '))   
    for data in data_list:
        news_data.append(data.split('/'))

news_data#取得所有文章的斷句
```

```python
char_list_and_tag=[]
tag=['B','I']
for cell in news_data:
    tag_read = cell[1]
    if tag_read=='nr':
        cell[1] = '-PER'
    elif tag_read=='ns':
        cell[1] = '-LOC'
    elif tag_read=='nt':
        cell[1] = '-ORG'
    elif tag_read=='o':
        cell[1] = 'O' 

    for index, char in enumerate(cell[0]):
        if index>0: index = 1
        if cell[1] == 'O': char_list_and_tag.append([char,cell[1]])
        else: char_list_and_tag.append([char,tag[index]+cell[1]]) 

char_list_and_tag[::-1]#倒過來看是否符合而已
```

```python
#儲存成taining data的格式
with open('../data_saved/train_data','a',encoding="utf-8-sig") as f:
    for item in char_list_and_tag:
        if item[0] =='!' or item[0] =='。':
            f.write(item[0]+"\t"+item[1]+"\n")
            f.write("\n")
        else : 
            f.write(item[0]+"\t"+item[1]+"\n")
    f.close
```
