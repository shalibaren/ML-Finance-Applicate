```python
import random
from collections import OrderedDict

news_data_list=[]

for i in range(1,8):
    news_data={}
    fp = open('./templates/news/'+str(i)+'.txt','r')# 有時候編碼有錯，要加,encoding="utf-8"    
    content = fp.read()
    content = content.replace("\n", '')    
    data_list = list(content.strip().split(' '))
    for data in data_list:
        data = data.split('/')
        news_data.update({data[0]: data[1]})
        news_data = OrderedDict(news_data)
    
    #print(news_data)
        
    org_list = (k for k,v in news_data.items() if v == 'nt')
    per_list = (k for k,v in news_data.items() if v == 'nr')
    loc_list = (k for k,v in news_data.items() if v == 'ns')   
    
    #print(org_list)
    
    n = open('./templates/names.txt','r')
    t_names = n.read()
    t_names = t_names.splitlines()
    
    o = open('./templates/orgs.txt','r')
    t_orgs = o.read()
    t_orgs = t_orgs.splitlines()
    
    l = open('./templates/locs.txt','r')
    t_locs = l.read()
    t_locs = t_locs.splitlines()
    
    
    for per in per_list:
        news_data = OrderedDict([(random.choice(t_names), v) if k == per else (k, v) for k, v in news_data.items()])
    for org in org_list:
        news_data = OrderedDict([(random.choice(t_orgs), v) if k == org else (k, v) for k, v in news_data.items()])
    for loc in loc_list:
        news_data = OrderedDict([(random.choice(t_locs), v) if k == loc else (k, v) for k, v in news_data.items()])
        
    
    for item in list(news_data.items()):
        news_data_list.append([item[0],item[1]])

news_data_list#取得所有文章的斷句
    

    
```

```python
char_list_and_tag=[]
tag=['B','I']
for cell in news_data_list:
    tag_read = cell[1]
    if tag_read=='nr':
        cell[1] = '-PER'
    elif tag_read=='ns':
        cell[1] = '-LOC'
    elif tag_read=='nt':
        cell[1] = '-ORG'
    elif tag_read=='o':
        cell[1] = 'O' 

    for index, char in enumerate(cell[0]):
        if index>0: index = 1
        if cell[1] == 'O': char_list_and_tag.append([char,cell[1]])
        else: char_list_and_tag.append([char,tag[index]+cell[1]]) 

char_list_and_tag[::-1]#倒過來看是否符合而已
```
