# Download dataset from 2000.07~2019.02

```python
import time
driver = webdriver.Chrome()
driver.get("https://www.sitca.org.tw/ROC/Industry/IN2201.aspx?pid=IN2221_01&fbclid=IwAR2cxosnAzJzr8bqHXJ56gwnqNMbRjavGt1BCCHjOVpUKiBKXVJMYA2GU3w")
time.sleep(10)
listmonth = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']
for i in range(2000, 2020):
    file_exist = True
    for j in listmonth:
        if i == 2000:
            if j not in ['07', '08', '09', '10', '11', '12']:
                file_exist = False
        elif i == 2019:
            if j not in ['01', '02']:
                file_exist = False
        if file_exist == True:
            try:
                driver.find_element_by_id("ctl00_ContentPlaceHolder1_ddlQ_YM").click()
                Select(driver.find_element_by_id("ctl00_ContentPlaceHolder1_ddlQ_YM")).select_by_visible_text(u"%s 年 %s 月" %(str(i), j))
                time.sleep(5)
                driver.find_element_by_id("ctl00_ContentPlaceHolder1_ddlQ_COLUMN").click()
                Select(driver.find_element_by_id("ctl00_ContentPlaceHolder1_ddlQ_COLUMN")).select_by_visible_text(u"定時定額")
                driver.find_element_by_id("ctl00_ContentPlaceHolder1_ddlQ_COLUMN").click()
                time.sleep(5)
                driver.find_element_by_id("ctl00_ContentPlaceHolder1_BtnQuery").click()
                time.sleep(5)
                driver.find_element_by_id("ctl00_ContentPlaceHolder1_BtnExport").click()
                time.sleep(6)
                print('file %s 年 %s 月 succesfully download' %(str(i), j))
            except:
                pass
```

```python
import os
path='E:\\user\\Documents\\GitHub\\ML-Finance-Applicate\\定期定額資料爬蟲\\' #這就是欲進行檔名更改的檔案路徑，路徑的斜線是為/，要留意下！
files=os.listdir(path)
print('files') #印出讀取到的檔名稱，用來確認自己是不是真的有讀到
names = []
for i in range(2001,2019):
    for j in range(1,13):
        names.append(str(i) + 'Y_' + str(j) + 'M')
names.append('2019Y_1M')
names.append('2019Y_2M')
n=0 #設定初始值
for i in files: #因為資料夾裡面的檔案都要重新更換名稱
    oldname=path+files[n] #指出檔案現在的路徑名稱，[n]表示第n個檔案
    newname=path+names[n]+'.xls' #在本案例中的命名規則為：年份+ - + 次序，最後一個.wav表示該檔案的型別
    os.rename(oldname,newname)
    print(oldname+'>>>'+newname) #印出原名與更名後的新名，可以進一步的確認每個檔案的新舊對應
    n=n+1 #當有不止一個檔案的時候，依次對每一個檔案進行上面的流程，直到更換完畢就會結束
    print(n) #確認檔案數量正確(218-1)
```
